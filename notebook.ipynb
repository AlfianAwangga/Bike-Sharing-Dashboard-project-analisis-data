#Exploratory Data Analysis
Pada proses pengolahan data ini, data akan diolah untuk menjawab pertanyaan berikut.
1. Apakah hari libur cenderung meningkatkan atau mengurangi jumlah penyewaan sepeda dibandingkan dengan hari kerja biasa?
2. Apakah tingkat angin dan kelembaban memengaruhi jumlah penyewaan sepeda?

Untuk menjawab pertanyaan #1, digunakan dataframe day_df.
Terlebih dahulu Menyiapkan dua buah dataframe (holiday_df dan workingday_df) untuk menyimpan data rata-rata penyewaan sepeda berdasarkan kondisi hari libur atau hari kerja

```python
holiday_df = day_df[day_df["holiday"] == 1]
holiday_df.head()
```

```python
workingday_df = day_df[day_df["workingday"] == 1]
workingday_df.head()
```

Menghitung data rata-rata dari total jumlah penyewaan pada hari libur dan hari kerja

```python
average_rent_on_holiday = holiday_df.cnt.mean()
print("Rata-rata penyewaan sepeda pada hari libur: ", average_rent_on_holiday)
```

```python
average_rent_on_workingday = workingday_df.cnt.mean()
print("Rata-rata penyewaan sepeda pada hari kerja: ", average_rent_on_workingday)
```

Menggabungkan kedua data kedalam satu dataframe

```python
holiday_vs_workingday_mean_df = pd.DataFrame({
    "category": ["Holiday", "Workingday"],
    "rent_average": [average_rent_on_holiday, average_rent_on_workingday]
})

holiday_vs_workingday_mean_df.sort_values(by="rent_average", ascending=False, inplace=True)
holiday_vs_workingday_mean_df
```

Visualisasi Data dengan Bar Chart, didapatkan bahwa rata-rata jumlah penyewaan sepeda pada hari kerja masih lebih banyak dibandingkan pada hari libur.

```python
colors = ["blue", "red"]

sns.barplot(
    x="category",
    y="rent_average",
    data=holiday_vs_workingday_mean_df,
    palette=colors
)
plt.title("Average Bike Rental on Workingday vs Holiday", fontsize=15)
plt.xlabel(None)
plt.ylabel(None)
plt.xticks(fontsize=12)
plt.show()
```

Untuk menjawab pertanyaan #2, digunakan dataframe hour_df untuk mendapatkan data angin dan kelembaban. Pertama-tama, membuat dua buah dataframe baru untuk mengelompokkan data jumlah penyewaan sepeda berdasarkan kecepatan angin dan kelembaban.

```python
by_windspeed_df = hour_df.groupby("windspeed").agg({
    "cnt": "sum"
})

by_windspeed_df.rename(columns={
    "cnt": "bike_rents"
}, inplace=True)

#mengkonversi satuan windspeed ke km/jam
by_windspeed_df["windspeed_kmh"] = by_windspeed_df.index*67

by_windspeed_df
```

```python
by_hum_df = hour_df.groupby("hum").agg({
    "cnt": "sum"
})

by_hum_df.rename(columns={
    "cnt": "bike_rents"
}, inplace=True)

#mengkonversikan satuan kelembaban menjadi %
by_hum_df["hum_percent"] = by_hum_df.index*100

by_hum_df
```

Selanjutnya, memvisualisasikan data yang telah didapat ke dalam bentuk line chart

```python
plt.figure(figsize=(10,5))

sns.lineplot(
    data=by_windspeed_df,
    x="windspeed_kmh",
    y="bike_rents"
)

plt.title("Correlation between Wind Speed and Number of Bicycle Rentals", fontsize=15)
plt.xlabel("Windspeed (Km/h)", fontsize=12)
plt.ylabel("Bike Rentals (Unit)", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.lineplot(
    data=by_hum_df,
    x="hum_percent",
    y="bike_rents",
    color="orange"
    
)

plt.title("Correlation between Humidity and Number of Bicycle Rentals", fontsize=15)
plt.xlabel("Humidity (%)", fontsize=12)
plt.ylabel("Bike Rentals (Unit)", fontsize=12)
plt.show()
```

Selanjutnya, mencoba untuk mengintepretasikan hasil data yang didapat berdasarkan korelasinya

```python
print("Korelasi antara kecepatan angin dan penyewaan sepeda: ",by_windspeed_df["windspeed_kmh"].corr(by_windspeed_df["bike_rents"]))
```

Korelasi antara kecepatan angin dan penyewaan sepeda menunjukkan nilai negatif. Ini dapat disimpulkan bahwa semakin meningkat kecepatan angin, semakin sedikit pula jumlah penyewaan sepeda

```python
print("Korelasi antara kelembaban dan penyewaan sepeda: ",by_hum_df["hum_percent"].corr(by_hum_df["bike_rents"]))
```

Korelasi antara kelembaban dan penyewaan sepeda menunjukkan nilai positif yang lemah. Dapat artikan bahwa jika kelembaban meningkat, jumlah penyewaan sepeda akan ikut meningkat, tetapi tidak signifikan.
